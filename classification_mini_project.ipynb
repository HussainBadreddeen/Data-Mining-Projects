## Dataset Description

The data set includes information about:

- Customers who left within the last month – the column is called Churn
- Services that each customer has signed up for – phone, multiple lines, internet, online security, online backup, device protection, tech support, and streaming TV and movies
- Customer account information – how long they’ve been a customer, contract, payment method, paperless billing, monthly charges, and total charges
- Demographic info about customers – gender, age range, and if they have partners and dependents

Each row represents a customer, and each column represents customer attributes as described below.

| Column | Description|
|-|-|
|gender|Whether the customer is a male or a female|
|SeniorCitizen| Whether the customer is a senior citizen or not (1, 0)|
|Partner|Whether the customer has a partner or not (Yes, No)|
|Dependents|Whether the customer has dependents or not (Yes, No)|
|tenure| Number of months the customer has stayed with the company|
|PhoneService|Whether the customer has a phone service or not (Yes, No)|
|MultipleLines|Whether the customer has multiple lines or not (Yes, No, No phone service*)|
|InternetService|Customer’s internet service provider (DSL, Fiber optic, No)|
|OnlineSecurity|Whether the customer has online security or not (Yes, No, No internet service**)|
|OnlineBackup|Whether the customer has online backup or not (Yes, No, No internet service)|
|DeviceProtection|Whether the customer has device protection or not (Yes, No, No internet service)|
|TechSupport|Whether the customer has tech support or not (Yes, No, No internet service)|
|StreamingTV|Whether the customer has streaming TV or not (Yes, No, No internet service) |
|StreamingMovies|Whether the customer has streaming movies or not (Yes, No, No internet service)|
|Contract|The contract term of the customer (Month-to-month, One year, Two year)|
|PaperlessBilling|Whether the customer has paperless billing or not (Yes, No)|
|PaymentMethod|The customer’s payment method (Electronic check, Mailed check, Bank transfer (automatic), Credit card (automatic))|
|MonthlyCharges|The amount charged to the customer monthly|
|TotalCharges|The total amount charged to the customer|
|Churn|Whether the customer churned or not (Yes or No)|

\* No phone service means the customer's plan does not include calling (could be a data-only plan)

\*\* No internet service means the customer's plan does not include internet/data and so the addon package does not apply

## Data Cleaning

```python
#Monthly charges column has all values are all STRINGS which is wrong, they containg '$' symbol to reflect dollar and '/m' to reflect per month
#To fix this we remove all $ and /m from the values and convert the strings to float
#then we change the name of the column to reflect that it is the Monthly charges per month in dollars ($)
df['MonthlyCharges'] = df['MonthlyCharges'].str.replace('$', '', regex=False)
df['MonthlyCharges'] = df['MonthlyCharges'].str.replace('/m', '', regex=False)
df['MonthlyCharges'] = pd.to_numeric(df['MonthlyCharges'], errors='coerce')
df=df.rename(columns={'MonthlyCharges': 'Monthly charges per month ($)'})

print(df['Monthly charges per month ($)'].head(10)) #testing to make sure the values dont have $ and dont have /m

print(df.head(5))# making sure the column name was changed successfully
```

```python
#Removing the dollar sign in the column and type casting it to numeric.
df['TotalCharges'] = df['TotalCharges'].str.replace('$', '', regex=False)
df['TotalCharges'] = pd.to_numeric(df['TotalCharges'], errors='coerce')
df=df.rename(columns={'TotalCharges': 'Total Charges ($)'})

print(df['Total Charges ($)'].head(10)) #testing to make sure the values dont have $ and dont have /m

print(df.head(5))# making sure the column name was changed successfull
```

```python
# Removing ID Column since it has a unique value for each instance and it can be dropped to avoid skewining the results instead of removing valuable data of the 56 rows with missing Ids
print("-----------------------------------------------------------------------")
df = df.drop('customerID',axis=1)

```

```python
# Calculating TotalCharges for the 852 rows with missing total charges.
#Calculation is done by multiplying Monthly charges by tenure to get the total instead of using the mean. why? we do this as this is more accurate then using the mean
# as we have a way to calculate the total charges more accurately instead of taking the mean. this method is often called Logical imputation
print("Amount of values in Total Charges ($)", df["Total Charges ($)"].count())
print("-----------------------------------------------------------------------")
print("Total number of null values in Total Charges ($) before doing logical imputation:",df['Total Charges ($)'].isnull().sum())#checking amount of null values before imputation
print("-----------------------------------------------------------------------")
df['Total Charges ($)'] = df.apply(lambda row: row['tenure'] * row['Monthly charges per month ($)'] if pd.isnull(row['Total Charges ($)']) else row['Total Charges ($)'], axis=1)
print("Total number of  null values in Total Charges ($) after doing logical imputation:",df['Total Charges ($)'].isnull().sum())#checking amount of null values after imputation
print("-----------------------------------------------------------------------")
print("Amount of values in Total Charges ($)", df["Total Charges ($)"].count()) # making sure if the amount of total charges are 7032 records

df.head(10)
```

## Data Preparation for Modelling

```python
#Copy dataset to encode
df_encoded = df.copy()
```

```python
df_encoded['Contract'] = df_encoded['Contract'].cat.codes
df_encoded['Churn'] = df['Churn'].astype('category').cat.codes
```

```python
categorical_columns=['MultipleLines','OnlineSecurity','OnlineBackup','DeviceProtection','TechSupport','StreamingTV','StreamingMovies','PaymentMethod','InternetService','gender','Partner','Dependents','PhoneService','PaperlessBilling']
df_encoded = pd.get_dummies(df_encoded, columns=categorical_columns)
df_encoded.info()
```

We label encoded Contract attribute since it has a meaningful order among its variables, Month-to-month < One year < Two years. The label encoding assigns a label to each category according to their order.
We also label encoded Churn as it is our target variable and assigning labels with weight to Yes and No helps in classification models. No < Yes


As for the rest we OHE the attributes since they are nominal and dont have any meaningful orders. The OHE represents each category as a seperate binary feature.


```python
#Here we drop one of the categories of all attributes that have two categories ex. (Yes or No) since after encoding they will be highly correlated.
df_encoded.drop(['gender_Female','Partner_No','Dependents_No','PhoneService_No','PaperlessBilling_No'], axis='columns', inplace=True)
df_encoded.columns
```

```python
#Here we split the data 30% testing and 70% Training
from sklearn.model_selection import train_test_split

X = df_encoded.drop('Churn', axis=1)
y = df_encoded['Churn']

X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.3, random_state=42)
```

## Modelling

```python
from sklearn.tree import DecisionTreeClassifier

tree_clf = DecisionTreeClassifier(random_state=42)
tree_clf.fit(X_train, y_train)
```

## Evaluation

```python
from sklearn.metrics import accuracy_score, confusion_matrix, classification_report

def print_score(clf, X_train, y_train, X_test, y_test, train=True):
    if train:
        pred = clf.predict(X_train)
        clf_report = pd.DataFrame(classification_report(y_train, pred, output_dict=True))
        print("Train Result:")
        print("===============================================")
        print(f"Accuracy Score: {accuracy_score(y_train, pred) * 100:.2f}%")
        print("_______________________________________________")
        print(f"CLASSIFICATION REPORT:\n{clf_report}")
        print("_______________________________________________")
        print(f"Confusion Matrix: \n {confusion_matrix(y_train, pred)}\n")

    elif train==False:
        pred = clf.predict(X_test)
        clf_report = pd.DataFrame(classification_report(y_test, pred, output_dict=True))
        print("Test Result:")
        print("===============================================")
        print(f"Accuracy Score: {accuracy_score(y_test, pred) * 100:.2f}%")
        print("_______________________________________________")
        print(f"CLASSIFICATION REPORT:\n{clf_report}")
        print("_______________________________________________")
        print(f"Confusion Matrix: \n {confusion_matrix(y_test, pred)}\n")
```

```python
print_score(tree_clf, X_train, y_train, X_test, y_test, train=True)
print_score(tree_clf, X_train, y_train, X_test, y_test, train=False)
```

```python
from sklearn.metrics import confusion_matrix, ConfusionMatrixDisplay

cm = confusion_matrix(y_test, tree_clf.predict(X_test), labels=tree_clf.classes_,normalize='true')
disp = ConfusionMatrixDisplay(confusion_matrix=cm,display_labels=tree_clf.classes_)
disp.plot()
plt.grid(False)
```

```python
from sklearn.metrics import f1_score, precision_score, recall_score

pred = tree_clf.predict(X_test)

print('Precision: ',precision_score(y_test,pred))
print('=======================================')
print('Recall: ',recall_score(y_test,pred))
print('=======================================')
print('F1_score: ',f1_score(y_test,pred))
```

Since our target variable is those customers who churned and the classes are imbalanced we would only consider the percision, recall, and f1 score of class 1.

**Percision Score:**
The precision score represents the percentage of actual churned customers (positive cases) out of the total customers the model predicted as churned. A precision score of 49.6% indicates that 49.6% of the customers the model predicted as churned actually churned. On the other hand, the remaining 50.4% of the predicted churned customers were false positives, customers predicted to churn who, in reality, did not churn. This low precision score suggests that the business would spend resources trying to retain customers who are unlikely to churn. This could result in wasted expenses, such as unnecessary marketing efforts or retention incentives being directed toward customers who would have stayed regardless.

**Recall Score:**
The recall score represents the percentage of actual churned customers (positive cases) correctly identified by the model out of all customers who actually churned. A recall score of 52.8% means that the model successfully identified 52.8% of the total churned customers. However, the remaining 47.2% of churned customers were not identified (false negatives). This low recall score indicates that the model is missing a significant portion of actual churned customers. As a result, the business might overlook critical opportunities to retain customers who are most likely to churn or have already churned.

**f1 Score:**
The F1 score is the harmonic mean of precision and recall, balancing the trade-off between the two metrics. It provides a single measure of a model’s effectiveness in handling imbalanced classes. In this case, the F1 score of 51.2% indicates that the model’s performance in predicting churned customers is modest, with room for improvement in either or both precision (avoiding false positives) and recall (identifying more true positives). The F1 score reflects that the model is only moderately effective at distinguishing churned customers. This means that the business may face two challenges simultaneously: Wasting resources on customers who are unlikely to churn due to low precision, and Missing out on opportunities to retain actual churned customers due to low recall.
